### Displaying images using matplotlib and opencv

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
import os
```

```python
## first we check if the path really exists!
path = "noor_north.jpg"
os.path.exists(path)
```

```python
img = cv2.imread(path)
```

### problem when showing images in matplotlib
OpenCV by default reads images in BGR format. To solve that, cvtColor(image, flag) can be used to convert BGR2RGB, from opencv to matplotlib, respectively. 

```python
plt.imshow(img)
```

```python
img_fixed = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fixed)
```

### writing on the image
let's suppose, that we want to write some information on the image using opencv.
This can be done, using putText command

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img_fixed,text="North of Iran",org=(800,600),fontFace=font,fontScale=10,color=(100,0,255),thickness=4)
```

```python
plt.imshow(img_fixed)
```
