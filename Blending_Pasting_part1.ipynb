### deraling with different sizes
* something to note is that the images have different sizes.
* This can also be checked with the command img1.shape.
* So, first we will resize them, to make them equal

```python
img1 = cv2.resize(img1,(1500,1200))
img2 = cv2.resize(img2,(1500,1200))
```

```python
plt.imshow(img1)
```

### blending the images

```python
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.1,gamma=0)
```

### blending images of different sizes together

```python
img1 = cv2.imread("DATA/noor_north.jpg")
img1= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("DATA/no_copy.png")
img2= cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

* taking a chunk of the larger image.
* for this stage we need some numpy slicing.
* the offsets are gonna stand for where you actually overlay the image, basically where you actually start the image. 

```python

x_offset = 0
y_offset = 0
```

* basically your x_end is x_offset plus the length of x in the small image. 

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

* take the large image and grab some chunck of x and y and reasign it to the small image


```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
